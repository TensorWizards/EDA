**Welcome, data enthusiasts!**

The `election_results_2024.csv` dataset contains the following columns:

1. **Constituency**: Name of the electoral constituency.
2. **Const. No.**: Unique identifier number for the constituency.
3. **Leading Candidate**: Name of the winning candidate.
4. **Leading Party**: Party of the winning candidate.
5. **Trailing Candidate**: Name of the candidate who lost.
6. **Trailing Party**: Party of the losing candidate.
7. **Margin**: Vote difference between the leading and trailing candidates.
8. **Status**: Status of the result, typically "Result Declared".

### Possible Explorations

1. **Distribution of Margins**: Analyze the winning margins.
2. **Party Analysis**: Count and compare seats won by each party.
3. **Regional Patterns**: Explore geographic trends if more location data is available.
4. **Close Contests and Landslide Victories**: Identify closely contested and landslide win constituencies.

# <p style="background-color:#3B6978;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">Visualization</p>

```python
# Plotting the count of leading parties
plt.figure(figsize=(10, 6))
sns.countplot(x='Leading Party', data=df, order=leading_party_counts.index)
plt.title('Count of Leading Parties')
plt.xlabel('Leading Party')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Margin'], bins=10, kde=True)
plt.title('Distribution of Winning Margins')
plt.xlabel('Winning Margin')
plt.ylabel('Frequency')
plt.show()


```

```python
# Histogram of winning margins by leading and trailing parties
plt.figure(figsize=(12, 8))
sns.histplot(df, x='Margin', hue='Status', multiple='stack', kde=True, bins=10)
plt.title('Histogram of Winning Margins by Status')
plt.xlabel('Winning Margin')
plt.ylabel('Frequency')
plt.legend(title='Status')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming `df` is your DataFrame with columns 'Constituency', 'Margin', and 'Leading Party'

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x=range(len(df)), y='Margin', hue='Leading Party', cmap='viridis')
plt.title('Scatter Plot of Margins for all Constituencies')
plt.xlabel('Index')
plt.ylabel('Margin')
plt.legend(title='Leading Party', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
